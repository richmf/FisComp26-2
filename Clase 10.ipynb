```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def integra_h2(f,x):
    integral , h = 0 , x[1]-x[0]
    for i in range(1,len(x)-1):
        integral += f[i+1] + 22*f[i] + f[i-1]
    integral += 0.5*(17*f[0] + 7*f[1] + f[2] - f[3])
    integral += 0.5*(17*f[-1] + 7*f[-2] + f[-3] - f[-4])
    return integral*h/24

def int_Riem(f,x):
    integral , h = 0 , x[1]-x[0]
    for i in range(len(x)-1):
        integral += f[i]
    return integral*h

def integra_h2_impar(f,x):
    integral , h = 0 , x[1]-x[0]
    for i in range(1,len(x),2):
        integral += f[i+1] + 4*f[i] + f[i-1]
    return integral*h/3

def integra_h2_par(f,x):
    integral , h = 0 , x[1]-x[0]
    for i in range(2,len(x),2):
        integral += f[i+1] + 4*f[i] + f[i-1]
    integral += (23*f[0] + 26*f[1] - 4*f[2] + 6*f[3] - 3*f[4])/16
    return integral*h/3

def integra_h2_mejorado(f,x):
    if len(x)%2 == 0:
        integral = integra_h2_par(f,x)
    else:
        integral = integra_h2_impar(f,x)
    return integral
```

```python
def integra_h2_unificado(f,x):
    integral , h , N = 0 , x[1]-x[0] , len(x)
    if N%2 == 0:
        inicio = 2
        integral += (23*f[0] + 26*f[1] - 4*f[2] + 6*f[3] - 3*f[4])/16
    else:
        inicio = 1
    for i in range(inicio,N,2):
        integral += f[i+1] + 4*f[i] + f[i-1]
    return integral*h/3
```

```python
x = np.linspace(0,np.pi,50)
y = np.sin(x)
print('h = ', x[1]-x[0])
print('Integral h2 ',integra_h2(y,x))
print('Integral Riem ',int_Riem(y,x))
print('Integral h2 mejorado',integra_h2_mejorado(y,x))
print('Integral h2 unificado',integra_h2_unificado(y,x))
```
