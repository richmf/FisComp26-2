Campo elctrico de una carga es
$$
\vec{E} = \frac{1}{4 \pi \epsilon _0} \frac{q \hat{(\vec r - \vec r_q)}}{|\vec r - \vec r_q|^2}
$$

```python
def E_q(r,r_q,q):
    d32 = ((r[0]-r_q[0])**2 + (r[1]-r_q[1])**2 )**1.5
    return (q*(r[0]-r_q[0]))/d32 , (q*(r[1]-r_q[1]))/d32
```

```python
x,y = np.linspace(-2,2) , np.linspace(-2,2)
XX,YY = np.meshgrid(x,y)
N = np.sqrt(XX**2 + YY**2)
plt.quiver(XX,YY,YY/N,-XX/N)
plt.streamplot(XX,YY,YY/N,-XX/N)
plt.gca().set_aspect('equal')
plt.show()
```

```python
Ex,Ey = E_q([XX,YY],[0,0],1)
plt.quiver(XX,YY,Ex,Ey)
plt.gca().set_aspect('equal')
plt.show()
```

```python
r_q , q = [0,0] , 1
Ex,Ey = E_q([XX,YY],r_q,q)
plt.streamplot(XX,YY,Ex,Ey,density=2)
plt.plot(r_q[0],r_q[1],'o')
plt.gca().set_aspect('equal')
plt.show()
```

```python
r_q , q = [0,1] , 1
Ex,Ey = E_q([XX,YY],r_q,q)
r_q_n , q_n = [0,-1] , -1
Ex_neg,Ey_neg = E_q([XX,YY],r_q_n,q_n)
Et_x = Ex + Ex_neg
Et_y = Ey + Ey_neg
plt.streamplot(XX,YY,Et_x,Et_y,density=2)
plt.plot(r_q[0],r_q[1],'ro')
plt.plot(r_q_n[0],r_q_n[1],'go')
plt.gca().set_aspect('equal')
plt.show()
```

```python
def E_tot(r,r_qs,qs):
    Ex , Ey = np.zeros_like(r[0]) , np.zeros_like(r[1])
    for i in range(len(qs)):
        E = E_q(r,r_qs[i],qs[i]) #Me regresa 2 variables
        Ex = Ex + E[0]
        Ey = Ey + E[1]
    return Ex,Ey
```

```python
pos_qs = [[0,1],[0,-1]]
qs = [1,-1]
Ex , Ey = E_tot([XX,YY],pos_qs,qs)
plt.streamplot(XX,YY,Ex,Ey,density=2)
plt.show()
```
