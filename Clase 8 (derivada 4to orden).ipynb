```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def deriv_h4(f,x):
    df = np.zeros_like(f)
    h = x[1]-x[0]
    for i in range(2,len(f)-2):
        df[i] = f[i-2] - 8*f[i-1] + 8*f[i+1] - f[i+2]
    df = df/12
    df[0] = -(25/12)*f[0] + 4*f[1] - 3*f[2] + (4/3)*f[3] - 0.25*f[4]
    df[1] = -(25/12)*f[1] + 4*f[2] - 3*f[3] + (4/3)*f[4] - 0.25*f[5]
    df[-1] = (25/12)*f[-1] - 4*f[-2] + 3*f[-3] - (4/3)*f[-4] + 0.25*f[-5]
    df[-2] = (25/12)*f[-2] - 4*f[-3] + 3*f[-4] - (4/3)*f[-5] + 0.25*f[-6]
    return df/h
```

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
dy = np.cos(x)
dy_num_h4 = deriv_h4(y,x)
plt.plot(x,dy,label='Derivada analítica')
plt.plot(x,dy_num_h4,label='Derivada numérica $h^4$')
plt.legend()
plt.show()
```
